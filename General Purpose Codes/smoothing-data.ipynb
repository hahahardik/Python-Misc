# Smoothing of Data

Refer 'smooth.txt' for data
 
To make data more smooth we need every value to be substituted by the average of it and its two neighbors. For example, if he have the sequence of 5 values like this:

> 3 5 6 4 5

Then the second (i.e. 5) should be substituted by (3 + 5 + 6) / 3 = 4.66666666667,<br>
the third (i.e. 6) should be substituted by (5 + 6 + 4) / 3 = 5,<br>
the fourth (i.e. 4) should be substituted by (6 + 4 + 5) / 3 = 5.

By agreement, the first and the last values will remain unchanged.

```python
def smooth(nums):
    '''This algo smooths the data (to be plotted) into more representable manner'''
    
    # Argument nums contains list of data points
    
    #Empty list to store the smoothed values
    t = []
    
    #According to the algo, first value remains same
    t.append(nums[0])
    
    for i in range(1,len(nums)-1):
        avg = (nums[i-1]+nums[i]+nums[i+1])/3
        t.append(avg)
    
    #According to the algo, first value remains same
    t.append(nums[-1])
    
    return t
```

### Small example

```python
list_i = [32.6, 31.2, 35.2, 37.4, 44.9, 42.1, 44.1]
print(smooth(list_i))
```

### Relatively big example
