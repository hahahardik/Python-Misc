# 004 - Largest Palindrome

A palindromic number reads the same both ways. The largest palindrome made from the product of two 2-digit numbers is 9009 = 91 Ã— 99.

Find the largest palindrome made from the product of two 3-digit numbers.

### I have taken a case for any number of digits.

```python
def palindrome(digit):
    
    # Empty list to store palindromes
    palindromes = []
    
    for i in range(10**(digit-1), 10**digit):
        for j in range(10**(digit-1), 10**digit):
            # Converted to string to check it with its reverse
            if str(i*j) == str(i*j)[::-1]:
                palindromes.append(i*j)
            else:
                continue
                
    # Returning a set to remove multiple entries
    return set(palindromes)
```

```python
# Argument '2' gives all the palindrome numbers obtained from multiplication of any two '2-digit' numbers
twod_pals = palindrome(2)
print(twod_pals)
```

```python
# Max can be obtained as follows
max(twod_pals)
```

```python
#For 3 Digits
print(palindrome(3))
```

```python
#Max 3-digit palindrome
max(palindrome(3))
```
